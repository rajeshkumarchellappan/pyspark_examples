```python
from functools import reduce
```

```python
order_file=open('retail_db/order_items/part-00000','r') # opening the file
```

```python
type(order_file)
```

```python
order_file_str=order_file.read() # convert the content into str
#order_file_str
```

```python
order_file_split=order_file_str.splitlines()
order_file_split1=order_file_split
order_file_split1
```

```python
# filter order id 2
order_file_split_filter=filter(lambda x:int(x.split(',')[1])==2,order_file_split1)
filtered=list(order_file_split_filter)
filtered
```

```python
order_file_split_fil_map=map(lambda x: float(x.split(',')[4]),filtered)
```

```python
order_file_split_fil_map1=list(order_file_split_fil_map)
```

```python
order_file_split_fil_map_red=reduce(lambda x,y:x+y,order_file_split_fil_map1)
total_revenue=order_file_split_fil_map_red
total_revenue
```
