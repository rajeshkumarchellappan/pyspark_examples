```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
from pyspark.sql import functions as func
```

```python
from pyspark.sql import Row
```

```python
#create a Spark Session
spark=SparkSession.builder.appName('SparkSql').getOrCreate()
```

```python
people = spark.read.option("header", "true").option("inferSchema", "true").csv("file:///SparkCourse/fakefriends-header.csv")
```

```python
type(people)
```

```python
people.printSchema()
```

```python
people.collect()
```

```python
people.show(10)
```

```python
friendsByAge=people.select('age','friends')
```

```python
friendsByAge.groupBy('age').avg('friends').show()
```

```python
friendsByAge.groupBy('age').avg('friends').sort('age').show()
```

```python
friendsByAge.groupBy('age').agg(func.round(func.avg('friends'),2)).sort('age').show()
```

```python
friendsByAge.groupBy('age').agg(func.round(func.avg('friends'),2).alias('Avg_Friends')).sort('age').show()
```
