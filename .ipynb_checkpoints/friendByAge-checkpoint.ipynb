```python
from pyspark import SparkConf,SparkContext
import collections
```

```python
conf=SparkConf().setMaster('local').setAppName('fakefriends')
sc=SparkContext(conf=conf)
lines=sc.textFile("file:///SparkCourse/fakefriends.csv")
print(lines.collect())
```

```python
def parseline(text):
    fields=text.split(',')
    age=int(fields[2])
    friends=int(fields[3])
    return (age,friends)
```

```python
age_friends=lines.map(parseline)
print(age_friends)
#averageFriendsByAge=age_friends.reduceByKey()
#print(averageFriendsByAge)
```

```python
totalFriendsByAge=age_friends.mapValues(lambda x:(x,1)).reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1]))
print(totalFriendsByAge.collect())
```

```python
averageFriendsByAge=totalFriendsByAge.mapValues(lambda x: x[0]/x[1])
print(averageFriendsByAge.collect())
```

```python
results=averageFriendsByAge.collect()
for r,r1 in results:
    print(r,r1)
```
