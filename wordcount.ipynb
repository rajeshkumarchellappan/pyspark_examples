```python
from pyspark import SparkConf,SparkContext
import collections
import re
conf=SparkConf().setMaster('local').setAppName('wordcount')
sc=SparkContext(conf=conf)
```

```python
rdd=sc.textFile('file:///sparkcourse/Book')
#rdd1=rdd.collect()
```

```python
def word_extract(text):
    #return re.split(r'[A-Za-z]', text.lower())
    return re.findall(r'[A-Za-z]+', text)

    
    
    #return re.findall(r'[A-Za-z]+',text)
```

```python
#re.compile(r'\W+', re.UNICODE).split(text.lower())
```

```python
rdd2=rdd.flatMap(word_extract)
wordCounts = rdd2.countByValue()
sort_wordCounts=collections.OrderedDict(sorted(wordCounts.items()))
for word, count in sort_wordCounts.items():
    #cleanWord = word.encode('ascii', 'ignore')
    #if (cleanWord):
     #   print(cleanWord.decode() + " " + str(count))
    print(word,count)
```
